ТОЕРИЯ: scatter plot

**Корреляция Пирсона** - степень линейной зависимости между двумя переменными 
1. Обратная линейная зависимость - при увеличении одной переменной другая уменьшается
2. Прямая линейная зависимость - при увеличении одной переменной другая увеличивается

- *Выбросы сильно влияют Корр. Пирсона*
- *Корр. Пирсона не подразумевает причинно-следственной связи*

|r| =< 0.3 - обратная высокая 
0.3 < |r| < 0.7 - умеренная  
|r| >= 0.7 - прямая высокая

При этом: 0.1 > |r| - взаимосвязи не наблюдается

**Корреляция Спирмена** -  мера монотонной (возрастающей или убывающей) взаимосвязи между двумя величинами
1. Основана не на самих значениях, а на рангах этих значений.
2. Изменяется от -1 до 1
3. Положительный коэффициент корреляции говорит о том, что с ростом одного признака возрастает и другой (монотонная прямая зависимость)
4. Отрицательный коэффициент корреляции говорит о том, что с ростом одного признака другой убывает (монотонная обратная зависимость)
5. Если корреляция равна 0 (или близка к 0), то монотонной зависимости между признаками нет
6. Корреляция Спирмана устойчива к выбросам, так как работает с рангами, а не с исходными значениями
7. Подходит для нелинейных, но монотонных зависимостей.
8. Также, как и корреляция Пирсона, не подразумевает причинно-следственной связи.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import numpy as np
np.set_printoptions(suppress=True, precision=2)
```

```python
df = pd.read_csv("/Users/nikolya/SnowFlake/Python_Analysis/Pandas_practice/Practice4/car_cell.csv")
```

```python
df.shape[0] - df.drop_duplicates().shape[0]
df = df.drop_duplicates()
```

```python
sns.boxplot(data=df, x="selling_price", y="owner")
sns.boxplot(data=df, x="selling_price", y="fuel")
```

```python
df[df["fuel"] == "petrol"]["selling_price"].mean() < df[df["fuel"] == "petrol"]["selling_price"].median() 
```

```python
df_encoded = pd.get_dummies(df, columns=["fuel", "owner"], dtype="int")
```

```python
df_encoded.corr(method="spearman", numeric_only=True).round(2)
```

```python
sns.violinplot(data=df_encoded, x="selling_price", y="seller_type")
sns.violinplot(data=df_encoded, x="selling_price", y="transmission")
```

```python
from sklearn.preprocessing import LabelEncoder as le

df_encoded["seller_type_encoded"] = le().fit_transform(df_encoded["seller_type"])
df_encoded["transmission_encoded"] = le().fit_transform(df_encoded["transmission"])

df_encoded["seller_type_encoded"].corr(df_encoded["selling_price"], method="spearman").round(2)
df_encoded["transmission_encoded"].corr(df_encoded["selling_price"], method="spearman").round(2)
```

```python
df_encoded.plot("km_driven", "selling_price", kind="scatter")
```

```python
df_encoded["selling_price"].corr(df_encoded["year"]).round(2) #Между годом выпуска (year) и ценой автомобиля (selling_price) существует ЗАМЕТНАЯ прямая линейная взаимосвязь
df_encoded["selling_price"].corr(df_encoded["km_driven"]).round(2)
```

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
df_encoded.corr(method="spearman", numeric_only=True).round(2)
df_encoded
```

```python
X1 = df_encoded[["fuel_Petrol", "owner_First Owner", "transmission_encoded", "year", "km_driven"]]
Y1 = df_encoded["selling_price"]

X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, Y1, test_size=0.2, random_state=42)
```

```python
y_test1.mean().round(2)
```

```python
model1 = LinearRegression().fit(X_train1, y_train1)
```

```python
model1.intercept_ # свободный коэффициент
```

```python
model1.coef_ # коэффициенты независимых переменных
```

```python
model1.predict([[0, 1, df_encoded["transmission_encoded"].mode()[0], 
                df_encoded["year"].quantile(0.25), df_encoded["km_driven"].median()]])
```

```python
y_pred1 = model1.predict(X_test1)

MSE1 = mean_squared_error(y_test1, y_pred1)
MAE1 = mean_absolute_error(y_test1, y_pred1)
R2_1 = r2_score(y_test1, y_pred1)

for i in (MSE1, MAE1, R2_1):
    print(round(i, 2))
```
